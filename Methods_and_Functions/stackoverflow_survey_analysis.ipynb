```python
completed_survey = new_survey['Country'].value_counts().sort_values(ascending= False).head(10)
completed_survey
```

```python
total = survey['Country'].value_counts()
total
```

```python
completed_country = [country for country in completed_survey.index]
completed_country
```

```python
total.loc['Italy']
```

```python
total_per_country = []
for item in completed_country:
    total_per_country.append(total.loc[item])
total_per_country
```

```python
percentage = pd.DataFrame()
percentage['Country'] = completed_country
percentage['Total'] = total_per_country
percentage['Completed'] = [item for item in completed_survey]
percentage.set_index('Country',inplace = True)

percentage
```

```python
percentage['%'] = round(percentage['Completed'] / percentage['Total'] *100,2)
percentage
```
